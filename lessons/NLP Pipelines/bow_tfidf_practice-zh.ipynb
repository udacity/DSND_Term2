# 词袋和 TF-IDF
接下来，我们将看到三种向量化文本的方法。
- `CountVectorizer` - 词袋 (Bag of Words)
- `TfidfTransformer` - TF-IDF 值
- `TfidfVectorizer` - 词袋和 TF-IDF 值

我们拿之前的示例数据，应用之前课上见过的文字处理方法看看效果。

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
corpus = ["The first time you see The Second Renaissance it may look boring.",
        "Look at it at least twice and definitely watch part 2.",
        "It will change your view of the matrix.",
        "Are the human people the ones who started the war?",
        "Is AI a bad thing ?"]
```

```python
stop_words = stopwords.words("english")
lemmatizer = WordNetLemmatizer()
```

使用你目前学习到的技术创建一个函数  `tokenize` ，接收一串文本，并做下列操作：
- 大小写标准化 (case normalization)，全部转换成小写
- 删除标点符号
- 分词 (tokenization)、词形还原（lemmatization）并使用 `nltk` 删除停止词

可以参考之前的章节完成以下步骤。

```python
def tokenize(text):
    # normalize case and remove punctuation
    text = re.sub(r"[^a-zA-Z0-9]", " ", text.lower())
    
    # tokenize text
    tokens = word_tokenize(text)
    
    # lemmatize andremove stop words
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]

    return tokens
```

# `CountVectorizer` (词袋)

```python
from sklearn.feature_extraction.text import CountVectorizer

# initialize count vectorizer object
vect = CountVectorizer(tokenizer=tokenize)
```

```python
# get counts of each token (word) in text data
X = vect.fit_transform(corpus)
```

```python
# convert sparse matrix to numpy array to view
X.toarray()
```

```python
# view token vocabulary and counts
vect.vocabulary_
```

# `TfidfTransformer`

```python
from sklearn.feature_extraction.text import TfidfTransformer

# initialize tf-idf transformer object
transformer = TfidfTransformer(smooth_idf=False)
```

```python
# use counts from count vectorizer results to compute tf-idf values
tfidf = transformer.fit_transform(X)
```

```python
# convert sparse matrix to numpy array to view
tfidf.toarray()
```

# `TfidfVectorizer`
`TfidfVectorizer` = `CountVectorizer` + `TfidfTransformer`

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# initialize tf-idf vectorizer object
vectorizer = TfidfVectorizer()
```

```python
# compute bag of word counts and tf-idf values
X = vectorizer.fit_transform(corpus)
```

```python
# convert sparse matrix to numpy array to view
X.toarray()
```
